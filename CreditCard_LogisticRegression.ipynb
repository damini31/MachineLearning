```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('creditcard.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
#Missing data check
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
#Imbalance data check
data['Class'].value_counts()
```

```python
x = data.iloc[:,0:-1]
x.head()
```

```python
y = data.iloc[:,-1]
y.head()
```

```python
#SMOTE for imbalance treatment
```

```python
from imblearn.over_sampling import RandomOverSampler
over = RandomOverSampler()
x_over, y_over = over.fit_resample(x,y)
```

```python
x_over.head()
```

```python
y_over.value_counts()
```

```python
x_over.shape
```

```python
y_over.shape
```

```python
#train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_over,y_over,test_size = 0.25,random_state = 101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LogitR = LogisticRegression()
LogitR.fit(x_train,y_train)
```

```python
y_pred = LogitR.predict(x_test)
```

```python
y_pred_train = LogitR.predict(x_train)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
print("the calssification report for train data :", classification_report(y_train,y_pred_train))
```

```python
print("the calssification report for test data :", classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import roc_auc_score
logit_roc_auc = roc_auc_score(y_test, y_pred)
logit_roc_auc
```

```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
display(fpr[:10])
display(tpr[:10])
display(thresholds[:10])
```

```python
#Plotting ROC & AUC Curve
```

```python
plt.figure(figsize=(15,15))
plt.plot(fpr, tpr, label="ROC and AUC curve (area=%0.2f)" % logit_roc_auc)
plt.plot([0,1],[0,1], 'k--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title("************Receiver Operating Characteristic with Area Under Curve************")
plt.legend(loc='lower right')
plt.show()
```
